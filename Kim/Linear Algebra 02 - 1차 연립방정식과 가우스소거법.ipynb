## Gauss Elimination
##### 순차적으로 문자(변수)를 앞에서부터 소거해준다.
##### pivot(non-zero) - 변수를 소거할때 기준으로 삼는 계수
##### all pivots are non-zero => G.E has a unique solution

1) pivot = 2
------------
2*u+v+w=5

4*u-6*v=-2

-2*u+7*v+2*w=9

2) pivot = -8
---------------
2*u+v+w=5

-8*v-2*w=-12

8*v+3*w=14

3) pivot = 1
-------------
2*u+v+w=5

w=2

4) 
------------------
w=2, v=1, u=1

##### when G.E has a unique solution, the last form of matrix is upper triangle


#### Breakdown 
- when a zero appears in a pivot position
- G.E has to stop
- the order of equation has to be changed(pivoting)


### Ex1

#### 1) 
-------------
u+v+w=a

2*u+2*v+5*w=b

4*u+6*v+8*w=c

#### 2) 
------------
u+v+w=a

3*w=b-2*a

2*v+4*w=c

(pivoting)

#### 3) 
--------------
u+v+w=a

2*v+4*w=c-4*a

3*w=b-2*a
