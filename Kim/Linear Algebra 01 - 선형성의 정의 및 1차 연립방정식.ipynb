Linearity 조건

1) Superposition, f(x1+x2) = f(x1) + f(x2)

2) Homogeniety, f(a*x) = a*f(x)

즉, f(a1*x2 + a2*x2) = a1*f(x1) + a2*f(x2)

------------------
example

y = m*x = f(x)

m(a1*x1 + a2*x2) = a1*m*x1 + a2*m*x2 = a1*f(x1) + a2*f(x2)


y = mx + n(!=0) = f(x)

m(a1*x1 + a2*x2) + n != a1*(m*x1 + n) + a2*(m*x2 + n)

이 경우에는 x와 y의 선형성은 없다
. 
(원점을 지나는 직선이 아니기 때문에)
